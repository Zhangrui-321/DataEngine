```python
import pandas as pd
from fbprophet import Prophet
# 读入数据集
df=pd.read_csv('train.csv')
df=df.set_index('ID')
df.head()
```

```python
new_col=['ds','y']
df.columns=new_col
m=Prophet()
m.fit(df)
```

```python
future=m.make_future_dataframe(periods=210)
#future.tail()
forecast = m.predict(future)
m.plot(forecast)
```

```python
m.plot_components(forecast)
```
