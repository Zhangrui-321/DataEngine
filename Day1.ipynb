```python
from pandas import Series,DataFrame
grades = {"语文":[68,95,98,90,80],'数学':[65,76,86,88,90],'英语':[30,98,88,77,90]}
df1=DataFrame(grades)
df2=DataFrame(grades,index=["张飞","关羽","刘备","典韦","许褚"])
print(df2)
#help(df2.describe)
mean_gr=df2.mean(1)
print("平均成绩为：",mean_gr)
min_gr=df2.min(1)
print("最低成绩为：",min_gr)
max_gr=df2.max(1)
print("最高成绩为：",max_gr)
var_gr=df2.var(1)
print("成绩方差为：",var_gr)
std_gr=df2.std(1)
print("成绩标准差为：",std_gr)
```

```python
import pandas as pd
sum_gr=df2.sum(1)
sum_gr=DataFrame(sum_gr,columns=['sum'])
#print(sum_gr)

df4=pd.concat([df2,sum_gr],axis=1)
df4.sort_values(by=["sum"],ascending = False,inplace = False)
```

```python
df5= df2.merge(sum_gr, left_index=True, right_index=True, how='left')
df5
```
